## PCA - Demo 2

* Using the PCA algorithm in a use case with images.
* MNIST dataset, very common for examples

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA

from sklearn.datasets import load_digits

## %matplotlib inline
```

## Load the dataset and inspect some information

```python
digits = load_digits()
print(digits.data.shape)
print(digits.images.shape)
print(np.unique(digits.target, return_counts=True))
```

```python
plt.gray()
plt.imshow(digits.images[9])
```

```python
digits.images[9].shape
```

## Applying PCA

```python
pca = PCA(n_components=8)
pca.fit(digits.data)
```

```python
pca.components_.shape
```

## Evaluating Variance

```python
pc_labels = []

for i in range(pca.components_.shape[0]):
    pc_labels.append('PC{}'.format(i+1))

print(pc_labels)
```

```python
plt.rcParams['figure.figsize'] = (14, 8)

sns.set(font_scale=2)
```

```python
data = {'Variance':pca.explained_variance_ratio_, 'Principal Components': pc_labels}

df_variance = pd.DataFrame(data)
sns.barplot(x='Principal Components', y='Variance', data=df_variance, color='c');
```

## Using PCA with only 2 Principal Components

```python
pca = PCA(n_components=2)
digits_transformed = pca.fit_transform(digits.data)
print(digits_transformed.shape)
```

```python
df = pd.DataFrame(digits_transformed, columns=['x', 'y'])
df['label'] = digits.target
df.head()
```

## Plotting Results

```python
_ = sns.scatterplot(x='x', y='y', data=df, hue='label', palette='rainbow', legend='full')
```

# Conclusion:
---
In this last example, we decided to use only two principal components to demonstrate how poor explainability can be when using low-scoring principal components.
