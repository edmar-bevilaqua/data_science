# Polynomial Regression
---
What if your data is more complex than a straight line?  
Surprisingly, you can use a linear model to fit nonlinear data. A simple way to do this is to add powers of each feature as new features, then train a linear model on this extended set of features.  
This technique is called ***Polynomial Regression***.  
Let’s look at an example. First, let’s generate some nonlinear data, based on a simple *quadratic equation* -> $y = ax^2 + bx + c$

```python
import matplotlib.pyplot as plt

plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)
```

```python
import numpy as np

points = 100
X = 6 * np.random.rand(points, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.rand(points, 1)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(X, y, 'y.')
```

```python
from sklearn.preprocessing import PolynomialFeatures

polynomial_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = polynomial_features.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X=X_poly, y=y)
```

```python
lin_reg.intercept_, lin_reg.coef_
```

```python
X_new = np.linspace(-3, 3, 100).reshape(100, 1)
X_new_poly = polynomial_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)

plt.figure(figsize=(12, 6))
plt.plot(X, y, 'y.')
plt.plot(X_new, y_new, 'r-')
```
